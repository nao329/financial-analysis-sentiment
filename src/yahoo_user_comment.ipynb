### トークン化関数

```python
def janome_token(text):
    token_filters = [CompoundNounFilter()]
    analyzer = Analyzer(char_filters=char_filters, tokenizer=Tokenizer(), token_filters=token_filters)
    tokens = analyzer.analyze(text) 
    sent = ""
    for token in tokens:
        if token.part_of_speech.split(',')[0] in ["名詞"]:
            # 名詞だけをリストに追加する
            sent += token.surface + " " # 表層形
        # 動詞（の原型），形容詞，副詞もリストに加えたい場合は次の２行を有効にする
#         if token.part_of_speech.split(',')[0] in ["動詞","副詞","形容詞"]:
#             sent += token.base_form + " " # 原形
    return sent
```

### ネガポジ分析

```python
import pandas as pd
pndic = pd.read_csv("./negapoji_dict/pn.csv.m3.120408.trim",
                    encoding="utf-8",
                    names=["word_type_score"]
                    )
print(pndic)
```

```python
import numpy as np

# 語と感情値を抽出
pndic["split"] = pndic["word_type_score"].str.split()
pndic["split"]
pndic["word"] = pndic["split"].str.get(0)
pndic["score"] = pndic["split"].str.get(1)

# # dict型に変換
keys = pndic['word'].tolist()
values = pndic['score'].tolist()
negapoji_dic = dict(zip(keys, values))
print(values)
```

```python
import pandas as pd

def score_convert(word):
    if word == "e":
        return 0
    elif word == "p":
        return 1
    elif word == "n":
        return -1
    else:
        return 0
    

def analyze_sentiment(dics,s_code_list):
    score = 0
    count = 0
    s_code_count=0
    advice_dic={}
    for dic in dics:
        for key, value in dic.items():
            if key in negapoji_dic.keys():
                score += int(score_convert(negapoji_dic[key]))*int(value)
                count += value
        if count >0:
            average_score = score / count
            if average_score > 0.8:
                advice= "買い注意(過熱感大)"
            elif 0.4 < average_score <= 0.8:
                advice = "超短期買い(過熱感あり)"
            elif 0 < average_score <= 0.4:
                advice= "短期買い(やや過熱感あり)"
            elif -0.5 < average_score <= 0:
                advice ="注意して買い(過熱感なし)"
            elif -0.8 < average_score <= 0.5:
                advice ="中期買い(過熱感なし)"
            else:
                advice= "リバ狙い買い(過熱感なし)"
            advice_dic[s_code_list[s_code_count]]=[advice,average_score]
            s_code_count+=1

    return advice_dic
```

```python
from janome.tokenizer import Tokenizer
from janome.analyzer import Analyzer
from janome.charfilter import *
from janome.tokenfilter import *
from janome.tokenizer import Tokenizer

token_filters = [CompoundNounFilter(), POSKeepFilter(['名詞']), TokenCountFilter(sorted=True)]
a = Analyzer(char_filters=char_filters, tokenizer=Tokenizer(), token_filters=token_filters)

dic_list=[]
for text in text_list:
    count = 0
    dic = {}
    for token in a.analyze(text):
        count+=1
        if count <= 10:
            dic[token[0]]=token[-1]
        print(token)
    print(dic)
    dic_list.append(dic)
```

```python
import json
advice_dic = analyze_sentiment(dic_list,s_code_list)
print(advice_dic)
# advice_dic.to_csv()
file_path = "./advice/" + today + ".json"
# file_path = "./advice/" + "2024-04-06" + ".json"    
    
json_data = json.dumps(advice_dic)

with open(file_path, mode='w',encoding="utf-8") as f:
    f.write(json_data)
```
